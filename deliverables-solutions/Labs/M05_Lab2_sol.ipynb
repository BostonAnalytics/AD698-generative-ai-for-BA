## Step 2: Scoring Rubric

Use 1-5 scales:

- `relevance_score`
- `groundedness_score`
- `clarity_score`


```python
#| eval: false
def simple_overlap_score(answer: str, reference: str) -> float:
	a = set(answer.lower().split())
	r = set(reference.lower().split())
	return len(a & r) / max(1, len(r))

def heuristic_groundedness(answer: str) -> int:
	# Replace with LLM-judge later; this is a quick baseline for the lab.
	if "insufficient" in answer.lower() or "according to" in answer.lower():
		return 4
	return 3
```

## Step 3: Build Comparative Table


```python
#| eval: false
rows = []
for _, row in eval_df.iterrows():
	q = row["question"]
	ref = row["reference_answer"]

	systems = {
		"prompt_only": run_prompt_only(q),
		"rag": run_rag(q),
		"pretrain_only": run_pretrain_only(q),
	}

	for name, ans in systems.items():
		rows.append({
			"question": q,
			"system": name,
			"answer": ans,
			"relevance_score": round(simple_overlap_score(ans, ref) * 5, 2),
			"groundedness_score": heuristic_groundedness(ans),
		})

result_df = pd.DataFrame(rows)
print(result_df.head())
```

## Step 4: Aggregate and Interpret


```python
#| eval: false
summary = (
	result_df
	.groupby("system")[["relevance_score", "groundedness_score"]]
	.mean()
	.reset_index()
)

print(summary)
summary.to_csv("M05_lab2_eval_summary.csv", index=False)
```
