# Create and Reshape a Tensor

Create a tensor containing integers from `0` to `23` and reshape it to `(3, 2, 4)`.

Tasks:

1. Print the tensor.
2. Print rank and shape.
3. Print dtype.


```python
#| eval: false
# Your code here
raw = torch.arange(24)
t = raw.reshape(3, 2, 4)

print(t)
print("rank:", t.ndim)
print("shape:", t.shape)
print("dtype:", t.dtype)
```

# Indexing and Reductions

Using tensor `t` from Question 1:

1. Extract the first 2D block (`t[0]`).
2. Compute column sums of that block.
3. Compute row means of that block.


```python
#| eval: false
# Your code here
block = t[0]
print("block:\n", block)
print("column sums:", block.sum(dim=0))
print("row means:", block.float().mean(dim=1))
```

# Broadcasting

Create:

1. `m` as a `(3, 4)` tensor with values `0` to `11`.
2. `bias` as a `(4,)` tensor `[0.1, 0.2, 0.3, 0.4]`.

Then compute `m + bias` and explain in one sentence why this works.


```python
#| eval: false
# Your code here
m = torch.arange(12).reshape(3, 4).float()
bias = torch.tensor([0.1, 0.2, 0.3, 0.4])
out = m + bias

print("m:\n", m)
print("bias:", bias)
print("m + bias:\n", out)
```

# CPU and GPU Device Practice

Create a random tensor of shape `(1000, 1000)` and move it to GPU if available.

Tasks:

1. Print original and target devices.
2. If GPU is available, compute a matrix multiplication on GPU.
3. Print output tensor shape and device.


```python
#| eval: false
# Your code here
x = torch.randn(1000, 1000)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
x_dev = x.to(device)

print("x device:", x.device)
print("x_dev device:", x_dev.device)

y_dev = x_dev @ x_dev.T
print("y_dev shape:", y_dev.shape)
print("y_dev device:", y_dev.device)
```
