# Why a Quarto Portfolio?

Employers in data science, business analytics, and AI roles increasingly expect
candidates to demonstrate work through **live, reproducible artifacts** — not
just PDF reports or PowerPoint slides.

A Quarto site lets you:

| Feature | Benefit |
|---|---|
| Render notebooks to HTML | Share interactive results without explaining how to run code |
| Version-control everything | Show your Git history — an indirect proxy for rigor |
| Deploy with one command | GitHub Pages is free, fast, and reliable |
| Mix prose and code | Tell the story alongside the analysis |
| Reference your own work | Cite analyses in assignments like a real researcher |

The course website you are reading right now is a Quarto site deployed on
GitHub Pages — you are already using the output of this workflow.

# Create a New Repository on GitHub

1. Go to [github.com/new](https://github.com/new).
2. Name your repository — a good convention is `<yourname>-portfolio` or `<yourname>.github.io`.
3. Set visibility to **Public** (required for free GitHub Pages).
4. Check **"Add a README file"**.
5. Click **Create repository**.

> **Tip — User Site vs. Project Site**
>
> GitHub supports two types of Pages deployments:
>
> | Type | Repository name | URL |
> |---|---|---|
> | **User site** | `<username>.github.io` | `https://<username>.github.io` |
> | **Project site** | anything else | `https://<username>.github.io/<repo>/` |
>
> The user site can only be deployed once per account.
> For a portfolio, either works — but the user site gives the cleanest URL.

# Clone the Repository Locally

```bash
git clone https://github.com/<your-username>/<your-repo>.git
cd <your-repo>
```

---

# Create a Quarto Project

Inside the cloned folder, initialise a Quarto website project:

```bash
quarto create project website .
```

Quarto will scaffold several files. Accept the defaults. You should now have:

```
_quarto.yml       ← site configuration
index.qmd         ← home page
about.qmd         ← about page (optional)
styles.css        ← custom CSS (optional)
```

---

# Configure `_quarto.yml`

`_quarto.yml` is the central configuration file for the entire site.
Open it and update it to match the structure below:

```yaml
project:
  type: website
  output-dir: _site

website:
  title: "Your Name — AI & Analytics Portfolio"
  description: "Projects in generative AI, NLP, and business analytics."
  navbar:
    left:
      - href: index.qmd
        text: Home
      - href: projects.qmd
        text: Projects
      - href: about.qmd
        text: About
  page-footer:
    left: "© 2025 Your Name"

format:
  html:
    theme: cosmo
    css: styles.css
    toc: true
    toc-depth: 2

execute:
  freeze: auto
  echo: true
  eval: true
```

Key fields explained:

| Field | Purpose |
|---|---|
| `output-dir: _site` | Where rendered HTML goes — this is what GitHub Pages serves |
| `navbar` | Top navigation links |
| `theme: cosmo` | Built-in Bootstrap theme (cosmo, flatly, darkly, etc.) |
| `freeze: auto` | Cache cell output; only re-execute cells whose source changed |

# Write Your Home Page (`index.qmd`)

Replace the contents of `index.qmd` with something like:

```markdown
---
title: "Welcome"
---

I am a Business Analytics student at Boston University studying **generative AI**,
**NLP**, and data-driven decision making.

## Featured Projects

- [SEC 10-K Word Embeddings](projects/m1-lab2.html) — Skip-gram embeddings
  trained on SEC filings to detect risk language similarity.
- [MLP Text Classifier](projects/m1-lab1.html) — TF-IDF + PyTorch pipeline for
  financial sentiment classification.
```

# Add a Project Page

Create a `projects/` folder and add a page for each lab:

```bash
mkdir projects
```

Copy a finished lab notebook (e.g. `M01_Lab2.qmd`) into `projects/` and add a
front-matter link back to the home page:

```yaml
---
title: "Lab 2 — Word Embeddings"
date: today
categories: [NLP, PyTorch, Embeddings]
---
```

Then register it in `_quarto.yml` under `navbar` or in a listing page.

---

# Render Locally

Before pushing to GitHub, always render and review locally:

```bash
quarto render
```

This places all output HTML in `_site/`.
Open `_site/index.html` in a browser to preview.

To render and instantly preview with live reload:

```bash
quarto preview
```

# Set Up GitHub Pages (docs/ method)

GitHub Pages can serve files from the `docs/` folder of the `main` branch.
This is the simplest method — no GitHub Actions required.

![GitHub Pages docs/ directory setting](./M01_lecture02_figures/gh-pages-docs-dir.png)

**Steps:**

1. In `_quarto.yml`, change `output-dir" to `docs`:

```yaml
project:
  type: website
  output-dir: docs
```

2. Re-render: `quarto render`

3. Commit and push:

```bash
git add .
git commit -m "Add Quarto site and initial portfolio content"
git push origin main
```

4. In your GitHub repository → **Settings** → **Pages**:
   - Source: **Deploy from a branch**
   - Branch: `main`
   - Folder: `/docs`
   - Click **Save**

![GitHub user site URL structure](./M01_lecture02_figures/gh-pages-user-site.png)

After 1–2 minutes, your site will be live at
`https://<username>.github.io/<repo>/` (project site) or
`https://<username>.github.io/` (user site).

# Recommended `.gitignore`

Add these lines to `.gitignore` so you do not commit unnecessary artifacts:

```
/.quarto/
*_cache/
*.quarto_ipynb
__pycache__/
.venv/
.env
```

Commit `docs/` but **not** `_site/` — `docs/` is the public output
that GitHub Pages serves.

---

# Update Workflow

Every time you finish a new lab or project:

1. Copy (or reference) the `.qmd` file into your portfolio project.
2. `quarto render`
3. `git add docs/ projects/ <changed-files>`
4. `git commit -m "Add M02 Lab 1 — LSTM language model"`
5. `git push origin main`

GitHub Pages re-deploys automatically within 60 seconds of the push.

# Portfolio Best Practices

| Practice | Rationale |
|---|---|
| Write a short narrative above each code cell | Demonstrates communication, not just coding |
| Include a results summary section | Recruiters scan, not read |
| Use `freeze: auto` | Keeps the repo lightweight; output is cached |
| Link each project from the home page | Makes navigation obvious |
| Add a one-paragraph "About" page with your LinkedIn | Enables professional contact |
| Use consistent date format (`date-format: long`) | Projects appear professional and current |
| Tag projects with categories | Enables filtering with Quarto listings |

# Adding a Quarto Listing Page

A **listing** automatically generates a gallery of all your project pages.
Add a file `projects.qmd`:

```yaml
---
title: "Projects"
listing:
  contents: projects
  sort: "date desc"
  type: grid
  categories: true
---
```

Quarto will scan all `.qmd` files in the `projects/` folder and generate
thumbnail cards with title, date, and category tags automatically.
