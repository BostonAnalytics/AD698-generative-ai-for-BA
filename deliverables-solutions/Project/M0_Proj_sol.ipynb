# What You Are Building

In this course project, you will build a domain-scoped financial RAG system over SEC 10-K filings.

Primary platform:
- Google Colab
- Google Drive storage

Optional platform:
- Local machine
- GitHub (optional only)

:::{.callout-important}
GitHub is not required. You will be graded on pipeline quality, grounding quality, and evaluation quality.
:::

# Four-Step Project Roadmap

1. Industry selection, SEC metadata parsing, 10-K download, item extraction.
2. Item selection, cleaning, and exploratory analysis.
3. Chunking, embeddings, and vector index construction.
4. RAG generation and evaluation.
