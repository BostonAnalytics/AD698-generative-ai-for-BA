# Presentation Guidelines

Each team presents their **financial technology product** — a domain-scoped RAG system built on SEC 10-K filings for a specific NAICS 4-digit industry. You are presenting to a simulated audience of FinTech stakeholders: potential investors, product managers, or a regulatory analytics team.

Each team gives a 10-12 minute presentation plus 3 minutes Q&A.

Expected flow:

1. NAICS industry selection and company scope — why this industry, who are the companies
2. Data pipeline and metadata quality — what you collected and how reliable it is
3. Item cleaning and EDA highlights — what patterns emerged in the corpus
4. Retrieval and generation pipeline — how the product answers questions
5. Evaluation results and failure cases — honest assessment of product performance
6. Practical recommendations — what would the next sprint look like for this product

# Required Evidence in Slides

- at least one data quality table
- at least one retrieval quality example
- at least one grounded answer with citations
- one evaluation summary chart
- one explainability or provenance slide showing how the team audited the system

## Recommended Explainability / Audit Slide

Because this is a RAG product, the most defensible explainability evidence is usually provenance rather than a generic "black box explanation." A strong slide can show one of the following:

- retrieved chunks with similarity or reranker scores
- answer text aligned to cited chunks
- failure case with explanation of why retrieval or generation failed
- SHAP/LIME/InterpretML output for an auxiliary classifier or market model, if the team built one

If teams use SHAP, LIME, or Microsoft explainability tools, they must clearly say **what model** is being explained. These tools are appropriate for a classifier, ranker, or tabular market model; they should not be described as a complete explanation of the generative model itself.
