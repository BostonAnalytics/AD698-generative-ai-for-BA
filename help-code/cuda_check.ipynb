```python
from manim import *

class Neuron(Scene):
    def construct(self):
        x = Text("x₁").shift(DOWN*2)
        h = Circle(radius=0.6).shift(UP*0.5)
        h_label = Text("h").move_to(h.get_center())
        y = Text("y").shift(UP*2)

        self.play(Write(x))
        self.play(Create(h), Write(h_label))
        self.play(Write(y))

        self.play(Create(Arrow(x.get_top(), h.get_bottom())))
        self.play(Create(Arrow(h.get_top(), y.get_bottom())))

        self.wait()
Neuron().render()
```
