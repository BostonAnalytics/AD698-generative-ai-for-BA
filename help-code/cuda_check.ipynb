```python
from manim import *

class Neuron(Scene):
    def construct(self):
        x = Text("x₁").shift(DOWN*2)
        h = Circle(radius=0.6).shift(UP*0.5)
        h_label = Text("h").move_to(h.get_center())
        y = Text("y").shift(UP*2)

        self.play(Write(x))
        self.play(Create(h), Write(h_label))
        self.play(Write(y))

        self.play(Create(Arrow(x.get_top(), h.get_bottom())))
        self.play(Create(Arrow(h.get_top(), y.get_bottom())))

        self.wait()
Neuron().render()
```

```python
from manim import *
from manim import config as global_config
from manim import WHITE

config = global_config.copy()
config.background_color = WHITE

class NeuronDiagram(Scene):
    def construct(self):
        # White background
        self.camera.background_color = WHITE

        # Colors
        input_color = BLUE
        neuron_color = MAROON
        text_color = BLACK

        # Neuron (Σ node)
        neuron = Circle(radius=0.6, color=neuron_color, stroke_width=1.5)
        neuron.move_to(ORIGIN)
        sigma = MathTex(r"\Sigma", color=neuron_color).scale(1.5)
        sigma.move_to(neuron.get_center())

        # Inputs X1, X2, X3
        x1 = MathTex(r"X_1", color=input_color).scale(0.9).move_to(2*LEFT + 1*UP)
        x2 = MathTex(r"X_2", color=input_color).scale(0.9).move_to(2*LEFT)
        x3 = MathTex(r"X_3", color=input_color).scale(0.9).move_to(2*LEFT + 1*DOWN)

        # Arrows from inputs to neuron
        a1 = Arrow(x1.get_right(), neuron.get_left() + 0.3*UP, buff=0.1, color=BLACK, stroke_width=1.5)
        a2 = Arrow(x2.get_right(), neuron.get_left(),              buff=0.1, color=BLACK, stroke_width=1.5)
        a3 = Arrow(x3.get_right(), neuron.get_left() + 0.3*DOWN,   buff=0.1, color=BLACK, stroke_width=1.5)

        # Bias
        bias = MathTex("b", color=text_color).move_to(1.2*UP)
        bias_arrow = Arrow(bias.get_bottom(), neuron.get_top(), buff=0.1, color=BLACK, stroke_width=1.5)

        # Output arrow + label
        out_arrow = Arrow(neuron.get_right(), neuron.get_right() + 2*RIGHT, buff=0.1, color=BLACK, stroke_width=1.5)
        out_label = MathTex("output", color=neuron_color).next_to(out_arrow, RIGHT)

        # Add everything to the scene (no animation, just static)
        self.add(
            neuron, sigma,
            x1, x2, x3,
            a1, a2, a3,
            bias, bias_arrow,
            out_arrow, out_label
        )
NeuronDiagram().render()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyArrowPatch
import numpy as np

def draw_neuron_vertical(path="neuron_vertical.svg"):
    plt.rcParams['font.family'] = 'serif'
    plt.rcParams['mathtext.fontset'] = 'cm'

    fig, ax = plt.subplots(figsize=(3, 4))
    ax.set_aspect('equal')
    ax.axis('off')

    # Neuron
    neuron_center = (0.5, 0.5)
    neuron = Circle(neuron_center, 0.10, edgecolor='maroon', facecolor='white', linewidth=2)
    ax.add_patch(neuron)
    ax.text(0.5, 0.5, r'$\Sigma$', ha='center', va='center', fontsize=16, color='maroon')

    # Input angles (in degrees)
    angles_deg = [65, 90, 115]
    labels = [r'$X_1$', r'$X_2$', r'$X_3$']

    for angle, lab in zip(angles_deg, labels):
        theta = np.deg2rad(angle)

        # Start point on a circle above the neuron
        r_start = 0.35
        x_start = neuron_center[0] + r_start * np.cos(theta)
        y_start = neuron_center[1] + r_start * np.sin(theta)

        # End point just above the neuron boundary
        r_end = 0.12
        x_end = neuron_center[0] + r_end * np.cos(theta)
        y_end = neuron_center[1] + r_end * np.sin(theta)

        # Label slightly above start
        ax.text(x_start, y_start + 0.03, lab,
                ha='center', va='bottom', fontsize=12, color='navy')

        # Arrow
        arrow = FancyArrowPatch((x_start, y_start), (x_end, y_end),
                                arrowstyle='->', mutation_scale=12,
                                linewidth=1.2, color='black')
        ax.add_patch(arrow)

    # Bias (left)
    ax.text(0.15, 0.5, r'$b$', ha='center', va='center', fontsize=12)
    ax.add_patch(FancyArrowPatch((0.22, 0.5), (0.40, 0.5),
                                 arrowstyle='->', mutation_scale=12,
                                 linewidth=1.2, color='black'))

    # Output (down)
    ax.add_patch(FancyArrowPatch((0.5, 0.38), (0.5, 0.12),
                                 arrowstyle='->', mutation_scale=12,
                                 linewidth=1.2, color='black'))
    ax.text(0.5, 0.08, 'output', ha='center', va='top', fontsize=12, color='maroon')

    fig.savefig(path, bbox_inches='tight', transparent=False)
    plt.close(fig)

draw_neuron_vertical("neuron_vertical.svg")
```

```python
from manim import *

class NeuralNetworkDiagram(Scene):
    def construct(self):
        # Layer sizes
        input_size = 8
        hidden_size = 4
        output_size = 2

        # Create layers
        input_layer = VGroup(*[
            Circle(radius=0.2, color=WHITE).shift(LEFT*4 + DOWN*(i - input_size/2)*0.6)
            for i in range(input_size)
        ])

        hidden_layer = VGroup(*[
            Circle(radius=0.2, color=WHITE).shift(DOWN*(i - hidden_size/2)*0.8)
            for i in range(hidden_size)
        ])

        output_layer = VGroup(*[
            Circle(radius=0.2, color=WHITE).shift(RIGHT*4 + DOWN*(i - output_size/2)*1.0)
            for i in range(output_size)
        ])

        # Add layers
        self.play(Create(input_layer), Create(hidden_layer), Create(output_layer))

        # Connect input → hidden
        for i_node in input_layer:
            for h_node in hidden_layer:
                self.play(Create(Line(i_node.get_right(), h_node.get_left())), run_time=0.05)

        # Connect hidden → output
        for h_node in hidden_layer:
            for o_node in output_layer:
                self.play(Create(Line(h_node.get_right(), o_node.get_left())), run_time=0.05)

        self.wait()
```

```python
import svgwrite
import os
import numpy as np
from IPython.display import SVG, display

def attention_output_svg(
    weights,
    input_labels=None,
    output_label="output",
    path="images/attention_output.svg",
    cmap_low="#ffffff",
    cmap_high="#cc0000",
    node_width=120,
    node_height=35,
    heat_width=40,
    x_spacing=180,
    y_spacing=60
):
    """
    weights: list or numpy array of attention weights (e.g., [0.1, 0.3, 0.6])
    input_labels: list of token labels (e.g., ["this", "movie", "great"])
    """

    os.makedirs(os.path.dirname(path), exist_ok=True)
    weights = np.array(weights)
    n = len(weights)

    if input_labels is None:
        input_labels = [f"x{i+1}" for i in range(n)]

    # Normalize weights for color mapping
    w_min, w_max = weights.min(), weights.max()
    norm = (weights - w_min) / (w_max - w_min + 1e-9)

    # SVG canvas
    dwg = svgwrite.Drawing(path, size=("700px", "500px"))

    # X positions
    x_input = 100
    x_heat = x_input + x_spacing
    x_output = x_heat + x_spacing

    # Y positions
    y_start = 250 - (n - 1) * y_spacing / 2

    # Draw input boxes
    input_positions = []
    for i, label in enumerate(input_labels):
        y = y_start + i * y_spacing
        dwg.add(dwg.rect(
            insert=(x_input - node_width/2, y - node_height/2),
            size=(node_width, node_height),
            fill="#D9ECFF",
            stroke="black",
            stroke_width=2
        ))
        dwg.add(dwg.text(
            label,
            insert=(x_input, y + 5),
            text_anchor="middle",
            font_size="16px",
            font_family="Arial"
        ))
        input_positions.append((x_input, y))

    # Draw heatmap column
    for i, w in enumerate(norm):
        y = y_start + i * y_spacing
        # interpolate color
        fill = svgwrite.utils.rgb(
            int((1-w)*255 + w*204),
            int((1-w)*255),
            int((1-w)*255),
            '%'
        )
        dwg.add(dwg.rect(
            insert=(x_heat - heat_width/2, y - node_height/2),
            size=(heat_width, node_height),
            fill=fill,
            stroke="black",
            stroke_width=1
        ))

    # Draw output box
    y_out = 250
    dwg.add(dwg.rect(
        insert=(x_output - node_width/2, y_out - node_height/2),
        size=(node_width, node_height),
        fill="#D9ECFF",
        stroke="black",
        stroke_width=2
    ))
    dwg.add(dwg.text(
        output_label,
        insert=(x_output, y_out + 5),
        text_anchor="middle",
        font_size="16px",
        font_family="Arial"
    ))

    # Draw weighted arrows
    for (x1, y1), w in zip(input_positions, weights):
        stroke_width = 1 + 6 * w  # thicker for larger weights
        dwg.add(dwg.line(
            start=(x1 + node_width/2, y1),
            end=(x_output - node_width/2, y_out),
            stroke="black",
            stroke_width=stroke_width,
            opacity=0.8
        ))

    dwg.save()
    return path

weights = [0.1, 0.3, 0.6]
input_labels = ["this", "movie", "great"]

image = attention_output_svg(weights, input_labels)

display(SVG(image))
```

```python
from pathlib import Path
import subprocess

out_dir = Path("../M0/M0_lecture03_figures")
out_dir.mkdir(exist_ok=True)

tex_file = out_dir / "backprop_tikz_demo.tex"

tikz_code = r"""
\documentclass[tikz,border=8pt]{standalone}
\usepackage{amsmath}
\usetikzlibrary{arrows.meta, positioning, calc, fit, backgrounds}

\definecolor{forward}{HTML}{2AA198}
\definecolor{backward}{HTML}{D95F5F}
\definecolor{nodeblue}{HTML}{3A86B7}
\definecolor{nodeorange}{HTML}{F28E2B}
\definecolor{lightgray}{HTML}{D9D9D9}

\begin{document}
\begin{tikzpicture}[
    neuron/.style={
        circle,
        draw=nodeblue,
        line width=1.2pt,
        minimum size=9mm,
        inner sep=1pt,
        fill=white,
        font=\small
    },
    active/.style={
        circle,
        draw=nodeorange,
        line width=1.6pt,
        minimum size=9mm,
        inner sep=1pt,
        fill=white,
        font=\small
    },
    faded/.style={
        circle,
        draw=lightgray,
        line width=1pt,
        minimum size=9mm,
        inner sep=1pt,
        fill=white,
        text=lightgray,
        font=\small
    },
    arrowf/.style={
        -{Latex[length=2.5mm]},
        line width=1.1pt,
        draw=forward
    },
    arrowb/.style={
        -{Latex[length=2.5mm]},
        line width=1.1pt,
        draw=backward
    },
    grayarrow/.style={
        -{Latex[length=2.2mm]},
        line width=.8pt,
        draw=lightgray
    },
    eqbox/.style={
        rounded corners=2mm,
        draw=backward,
        line width=.9pt,
        fill=white,
        align=left,
        inner sep=5pt,
        font=\scriptsize
    }
]

% ------------------------------------------------
% Layer coordinates
% ------------------------------------------------
\node[faded]  (x0) at (0,  1.4) {$x_0$};
\node[faded]  (x1) at (0,  0.2) {$x_1$};
\node[neuron] (x2) at (0, -1.0) {$x_2$};

\node[neuron] (h1a) at (2.3,  1.2) {};
\node[neuron] (h1b) at (2.3,  0.0) {$z|a$};
\node[neuron] (h1c) at (2.3, -1.2) {};

\node[neuron] (h2a) at (4.6,  1.2) {};
\node[neuron] (h2b) at (4.6,  0.0) {$z|a$};
\node[neuron] (h2c) at (4.6, -1.2) {};

\node[active] (out) at (6.9, 0) {$z|a$};

\node[font=\small] at (0,   2.2) {$(0)$};
\node[font=\small] at (2.3, 2.2) {$(L-2)$};
\node[font=\small] at (4.6, 2.2) {$(L-1)$};
\node[font=\small] at (6.9, 2.2) {$(L)$};

% ------------------------------------------------
% Faded dense connections
% ------------------------------------------------
\foreach \i in {x0,x1,x2}{
  \foreach \j in {h1a,h1b,h1c}{
    \draw[grayarrow] (\i) -- (\j);
  }
}

\foreach \i in {h1a,h1b,h1c}{
  \foreach \j in {h2a,h2b,h2c}{
    \draw[grayarrow] (\i) -- (\j);
  }
}

\foreach \i in {h2a,h2b,h2c}{
  \draw[grayarrow] (\i) -- (out);
}

% ------------------------------------------------
% Highlighted forward path
% ------------------------------------------------
\draw[arrowf] (x2) -- node[below, font=\scriptsize] {$w_3$} (h1b);
\draw[arrowf] (h1b) -- node[below, font=\scriptsize] {$w_2$} (h2b);
\draw[arrowf] (h2b) -- node[below, font=\scriptsize] {$w_1$} (out);

\node[font=\scriptsize, text=forward] at (2.3,-.72) {$\varphi$};
\node[font=\scriptsize, text=forward] at (4.6,-.72) {$\varphi$};
\node[font=\scriptsize, text=forward] at (6.9,-.72) {$\varphi$};

% ------------------------------------------------
% Output error
% ------------------------------------------------
\node[font=\small] (err) at (8.7, 0) {$E_2$};
\draw[arrowf] (out) -- node[above, font=\scriptsize] {$a^{(L)}-y$} (err);

% ------------------------------------------------
% Backpropagation arrows
% ------------------------------------------------
\draw[arrowb, bend right=18] (err.south) to node[below, font=\scriptsize] {$1$} (out.south);
\draw[arrowb, bend right=28] (out.south) to node[below, font=\scriptsize] {$2$} (h2b.south);
\draw[arrowb, bend right=38] (h2b.south) to node[below, font=\scriptsize] {$3$} (h1b.south);

\node[font=\scriptsize, text=backward] at (6.9,-1.05) {$\varphi'$};
\node[font=\scriptsize, text=backward] at (4.6,-1.05) {$\varphi'$};
\node[font=\scriptsize, text=backward] at (2.3,-1.05) {$\varphi'$};

% ------------------------------------------------
% Number markers
% ------------------------------------------------
\node[circle, draw=backward, fill=white, minimum size=6mm, font=\scriptsize] at (7.35,-.75) {1};
\node[circle, draw=backward, fill=white, minimum size=6mm, font=\scriptsize] at (4.95,-1.65) {2};
\node[circle, draw=backward, fill=white, minimum size=6mm, font=\scriptsize] at (2.55,-1.65) {3};

% ------------------------------------------------
% Equation callout boxes
% ------------------------------------------------
\node[eqbox] (eq1) at (8.2,-2.1) {
$ \displaystyle
\frac{\partial E_2}{\partial w_1}
=
a^{(L-1)} \cdot \varphi'(z^{(L)}) \cdot 1
$
};

\node[eqbox] (eq2) at (4.8,-3.05) {
$ \displaystyle
\frac{\partial E_2}{\partial w_2}
=
a^{(L-2)} \cdot \varphi'(z^{(L-1)})
\cdot w_1 \cdot \varphi'(z^{(L)}) \cdot 1
$
};

\node[eqbox] (eq3) at (1.9,-3.95) {
$ \displaystyle
\frac{\partial E_2}{\partial w_3}
=
a^{(L-3)} \cdot \varphi'(z^{(L-2)})
\cdot w_2 \cdot \varphi'(z^{(L-1)})
\cdot w_1 \cdot \varphi'(z^{(L)}) \cdot 1
$
};

% ------------------------------------------------
% Title
% ------------------------------------------------
\node[font=\bfseries] at (2.7, 3.0) {Gradient descent \textbar{} partial derivative};

\end{tikzpicture}
\end{document}
"""

tex_file.write_text(tikz_code, encoding="utf-8")

# Compile with latexmk if available
subprocess.run(
    ["latexmk", "-pdf", "-interaction=nonstopmode", tex_file.name],
    cwd=out_dir,
    check=False
)

print(f"Saved: {tex_file}")
print(f"Expected PDF: {tex_file.with_suffix('.pdf')}")
```
