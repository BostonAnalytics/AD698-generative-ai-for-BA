## 3. Tensor Basics

Create tensors, inspect shapes/dtypes, and run simple element-wise operations.

```python
x = torch.arange(0, 12).reshape(3, 4).float()
y = torch.ones((3, 4))

print('x:\n', x)
print('shape:', x.shape, 'dtype:', x.dtype)
print('x + y:\n', x + y)
print('row sums:', x.sum(dim=1))
print('column means:', x.mean(dim=0))
```

## 4. Move Tensors Between CPU and GPU

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
x_dev = x.to(device)

print('target device:', device)
print('x original device:', x.device)
print('x moved device:', x_dev.device)
```

## 5. Simple CPU vs GPU Benchmark

This benchmark uses matrix multiplication to compare runtime.

```python
import time

size = 2000
a_cpu = torch.randn(size, size)
b_cpu = torch.randn(size, size)

t0 = time.time()
_ = a_cpu @ b_cpu
cpu_s = time.time() - t0

if torch.cuda.is_available():
    a_gpu = a_cpu.cuda()
    b_gpu = b_cpu.cuda()
    torch.cuda.synchronize()
    t1 = time.time()
    _ = a_gpu @ b_gpu
    torch.cuda.synchronize()
    gpu_s = time.time() - t1
    print(f'CPU: {cpu_s:.4f}s | GPU: {gpu_s:.4f}s')
else:
    print(f'CPU: {cpu_s:.4f}s | GPU not available')
```
